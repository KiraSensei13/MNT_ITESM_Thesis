```python
#******************************************************************************#
# Futures
%matplotlib inline
# from __future__ import unicode_literals
# from __future__ import print_function

# Generic/Built-in
import datetime
import argparse

# Other Libs
from IPython.display import display, Image
#
from sympy import *
#
import matplotlib.pyplot as plt
plt.rc('xtick', labelsize=15)     
plt.rc('ytick', labelsize=15)
#
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
#
from pint import UnitRegistry
u = UnitRegistry()
u.default_format = 'P'

# Owned
pfont = {'fontname':'Consolas'};
# from nostalgia_util import log_utils
# from nostalgia_util import settings_util
__authors__   = ["Osamu Katagiri - A01212611@itesm.mx"]
__copyright__ = "None"
__credits__   = ["Marcelo Videa - mvidea@itesm.mx"]
__license__   = "None"
__status__    = "Under Work"
#******************************************************************************#

scale = 6;
fig, ax = plt.subplots(figsize=(3*scale, 2*scale));

# 1st pyrolysis ramp
# heat ramp of 5 C/min up to 300 C
t = np.linspace(0,300/5,1000);
T = (5/1)*t + 0;
plt.plot(t,T, '-', linewidth=3, label='heat ramp of ' + r'$5 C/min$' + ' up to ' + r'$300 C$');

t = np.linspace(300/5,(300/5)+60,1000);
T = np.linspace(300,300,1000);
plt.plot(t,T, '-', linewidth=3);

# 2nd pyrolysis ramp
# heat ramp of 5 C/min up to 900 C
t = np.linspace((300/5)+60,(300/5)+(900/5),1000);
T = (5/1)*t - 300;
plt.plot(t,T, '-', linewidth=3, label='heat ramp of ' + r'$5 C/min$' + ' up to ' + r'$900 C$');

t = np.linspace((300/5)+(900/5),(300/5)+(900/5)+60,1000);
T = np.linspace(900,900,1000);
plt.plot(t,T, '-', linewidth=3);

# 3rd pyrolysis ramp
# heat ramp of 5 C/min down to room temperature
t = np.linspace((300/5)+(900/5)+60,(300/5)+(900/5)+235,1000);
T = -(5/1)*t + 2400;
plt.plot(t,T, '-', linewidth=3, label='heat ramp of ' + r'$5 C/min$' + ' down to ' + str(min(T)) + r'$C$');

t = np.linspace(0,(300/5)+(900/5)+235,1000);
T = np.linspace(min(T),min(T),1000);
plt.plot(t,T, ':', linewidth=1.5, label='room temperature');

# Plot config
plt.xscale('linear')
plt.yscale('linear')
plt.xlabel(r'$t$' + '    ' + r'$[min]$', fontsize=24, **pfont);
plt.ylabel(r'$T$' + '    ' + r'$[C]$', fontsize=24, **pfont);
plt.title('Pyrolysis Process', size=24, **pfont);
#plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.legend(prop={'size': 18});
plt.annotate(r'$1 hr$', xy=(40,150), xycoords='data', fontsize=20, **pfont)
plt.annotate(r'$1 hr$', xy=(80,310), xycoords='data', fontsize=20, **pfont)
plt.annotate(r'$2 hrs$', xy=(190,600), xycoords='data', fontsize=20, **pfont)
plt.annotate(r'$1 hr$', xy=(260,910), xycoords='data', fontsize=20, **pfont)
plt.annotate(r'$\sim 3 hrs$', xy=(375,550), xycoords='data', fontsize=20, **pfont)
display(plt);
```
