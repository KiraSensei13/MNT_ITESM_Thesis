```python
# PYTHON LIBRARIES
%matplotlib inline

import numpy as np
np.seterr(divide='ignore', invalid='ignore')
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('xtick', labelsize=15)     
plt.rc('ytick', labelsize=15)

from scipy import special, optimize

# DATA FIG 1
alpha = [0.003652583, 0.001774215, 0.031965924, 0.075, 0.161258301, 0.526700194, 0.731638427, 0.8, 0.82, 0.83, 0.83];
rate = [0.019713504, 0.355176935, 0.79274908, 1.434469194, 2.875276684, 12.18139645, 6.831274434, 2.94538576, 1.092906364, 0.333333333, 0.333333333];
```

$$ rate = k \alpha^m {( 1 - \alpha )}^n $$

```python
#eq 21
def rate_(alpha, k, m, n):
    #res = k * (alpha**m) * ((1 - alpha)**n);
    res = [k * (a**m) * ((1 - a)**n) for a in alpha]
    return res
```

```python
# reasonable initial guesses for EOS parameters
k = 800
m = 3
n = 3

rate_prev = rate_(alpha, k, m, n);

p0 = k, m, n
results = optimize.curve_fit(rate_, alpha, rate, p0)

k = results[0][0];
m = results[0][1];
n = results[0][2];

rate_fit = rate_(alpha, k, m, n);
print(*results[0]);
```

```python
# PLOT FIG 1
scale = 6;
fig, ax = plt.subplots(figsize=(3*scale, 2*scale));

# Plot
plt.scatter(alpha, rate, s=25, color='red', label='Raw data');
alpha_space = np.linspace(min(alpha), max(alpha), 100)
rate_space = rate_(alpha_space, k, m, n)
plt.plot(alpha_space, rate_space, '-', linewidth=3, label='Fitted curve')

# Print fitting parameters as plot legends
x = alpha[0]
y = rate[0]
plt.scatter(x, y, s=0, label=r'$k = $' + str(round(k, 2)))
plt.scatter(x, y, s=0, label=r'$m = $' + str(round(m, 2)))
plt.scatter(x, y, s=0, label=r'$n = $' + str(round(n, 2)))

# Display plots
plt.yscale('linear');
plt.xlabel(r'$\alpha$', fontsize=24);
plt.ylabel(r'$rate$', fontsize=24);
plt.legend(prop={'size': 18});
display(plt);
```
