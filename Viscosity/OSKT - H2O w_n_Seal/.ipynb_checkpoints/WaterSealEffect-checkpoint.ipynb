```python
# PYTHON LIBRARIES
%matplotlib inline

import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt
plt.rc('xtick', labelsize=15)     
plt.rc('ytick', labelsize=15)
```

```python
def get_csv(filename):
    df = pd.read_csv(filename, delimiter=","); # Get CSV data
    units = df.iloc[0]; # Get units
    df = df.drop([0]); # Remove units row from dataframe
    
    #      [0]    [1]
    return units, df
```

```python
# Get units
units_df = get_csv("./oskt_h2onseal.csv")[0]
display(units_df)

# Get CSV data
nSeal_df = get_csv("./oskt_h2onseal.csv")[1]
wSeal_df = get_csv("./oskt_h2owseal.csv")[1]

scale = 6;
plt.subplots(figsize=(3*scale, 2*scale));

# Plot
x = nSeal_df["Shear rate"].to_numpy().astype(np.float)
y = nSeal_df["Viscosity"].to_numpy().astype(np.float)
#x = np.exp(x)
#y = np.exp(y)
plt.plot(x, y, linewidth=3, label='with no water seal');

x = wSeal_df["Shear rate"].to_numpy().astype(np.float)
y = wSeal_df["Viscosity"].to_numpy().astype(np.float)
#x = np.exp(x)
#y = np.exp(y)
plt.plot(x, y, linewidth=3, label='with water seal');

# Display plots
plt.yscale('log');
plt.xscale('log');
plt.xlabel(str(nSeal_df.columns[1]) + " [" + str(units_df[1]) + "]", fontsize=24);
plt.ylabel(str(nSeal_df.columns[2]) + " [" + str(units_df[2]) + "]", fontsize=24);
#plt.title('Figure 6.1', size=24);
plt.legend(prop={'size': 18});
display(plt);
```
